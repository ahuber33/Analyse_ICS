# Tracking Lysosomes — Suivi temporel par microscopie de fluorescence

Ce notebook orchestre le pipeline complet de **suivi temporel de lysosomes**
dans un stack TIFF multi-frames. Il s'appuie sur les modèles entraînés dans
`Generation_ML_puncta.ipynb` et sur les fonctions utilitaires de `Include.ipynb`.

**Pipeline global :**
1. Détection automatique GPU / CPU
2. Configuration de tous les paramètres
3. Chargement des modèles (`SomaCNN` + Cellpose `cyto3`)
4. Lecture et tri chronologique du stack TIFF
5. Suivi des lysosomes (TrackPy seul **ou** CNN + Cellpose + TrackPy)
6. Génération et sauvegarde de 7 figures d'analyse

**Deux modes de détection disponibles :**
- `flag_CNN=True` : pipeline complet CNN → Cellpose → TrackPy (plus précis)
- `flag_CNN=False` : TrackPy seul (plus rapide, moins précis)

## 5. Configuration des paramètres et chargement des modèles

Tous les paramètres du pipeline sont centralisés ici pour faciliter
la reproductibilité et les comparaisons entre expériences.

### Section 1 — Dispositif de calcul
`get_device()` détecte automatiquement GPU (CUDA) ou CPU et affiche
le matériel utilisé.

### Section 2 — Paramètres du pipeline

#### Prétraitement image
| Paramètre | Défaut | Description |
|---|---|---|
| `pretraitement_sigma` | 1.5 | Sigma du flou gaussien (px) |
| `clip_limit` | 0.02 | Limite CLAHE ∈ [0, 1] |

#### CNN soma
| Paramètre | Défaut | Description |
|---|---|---|
| `cnn_threshold` | 0.23 | Seuil de probabilité CNN (↓ = plus de détections) |
| `patch_size` | 64 | Taille du patch carré (px) — doit correspondre à l'entraînement |
| `cnn_batch_size` | 512 | Patches par passe CNN (↑ si VRAM ≥ 6 Go, ↓ si erreur OOM) |
| `overlap_thresh` | 0.1 | Fraction de chevauchement pour supprimer une ROI redondante |

#### Cellpose (lysosomes)
| Paramètre | Défaut | Description |
|---|---|---|
| `diameter_cellpose` | 11 | Diamètre estimé des lysosomes (px) |
| `cellprob_threshold` | 0.1 | Seuil probabilité Cellpose |
| `flow_threshold` | 0.5 | Cohérence des flux optiques |
| `min_area` / `max_area` | 5 / 500 | Filtre aire (px²) |
| `min_circularity` | 0.5 | Circularité minimale (filtre artefacts) |
| `max_axis_ratio` | 2 | Rapport axes max (filtre dendrites) |
| `top_hat_radius` | 10 | Rayon disque top-hat (> rayon lysosome) |

#### TrackPy
| Paramètre | Défaut | Description |
|---|---|---|
| `diameter_trackpy` | 11 | Diamètre particules pour `tp.locate` (doit être impair) |
| `search_range` | 4 | Portée de recherche entre frames (px) — mettre 2 avec CNN+Cellpose |
| `memory` | 4 | Frames de mémorisation avant abandon d'une trajectoire |
| `threshold_filtered` | 20 | Longueur minimale de trajectoire conservée (frames) |

### Section 3 — Chargement des modèles
Les modèles sont chargés **une seule fois** sur le bon dispositif,
ce qui évite les transferts répétés CPU↔GPU pendant la boucle d'analyse.


```python
# ═══════════════════════════════════════════════════════════════════════════
# SECTION 1 — DÉTECTION DU DISPOSITIF DE CALCUL
# ═══════════════════════════════════════════════════════════════════════════

# get_device() détecte automatiquement CUDA ou CPU et affiche le résultat
device = get_device()


# ═══════════════════════════════════════════════════════════════════════════
# SECTION 2 — PARAMÈTRES DU PIPELINE
# ═══════════════════════════════════════════════════════════════════════════

# ── Prétraitement image ───────────────────────────────────────────────────
pretraitement_sigma = 1.5   # écart-type du flou gaussien (px)
clip_limit          = 0.02  # limite CLAHE ∈ [0, 1]

# ── CNN de détection de somas ─────────────────────────────────────────────
cnn_threshold  = 0.23   # 🔧 seuil probabilité CNN (calibré via courbe PR)
patch_size     = 64     # taille du patch carré (doit correspondre à l'entraînement)
half           = patch_size // 2
cnn_batch_size = 512    # 🔧 ↑ si VRAM ≥ 6 Go, ↓ si erreur OOM GPU

# ── Fusion des ROI ────────────────────────────────────────────────────────
overlap_thresh = 0.1    # fraction chevauchement min pour supprimer une ROI

# ── Segmentation Cellpose (lysosomes) ────────────────────────────────────
diameter_cellpose  = 11    # 🔧 diamètre estimé lysosomes (px)
cellprob_threshold = 0.1   # seuil Cellpose (↓ = plus de détections)
flow_threshold     = 0.5   # cohérence flux (↑ = plus strict)
min_area           = 5     # 🔧 aire minimale lysosome (px²)
max_area           = 500   # 🔧 aire maximale lysosome (px²)
min_circularity    = 0.5   # circularité min (filtre artefacts allongés)
max_axis_ratio     = 2     # rapport axes max (filtre dendrites)
top_hat_radius     = 10    # rayon disque top-hat (doit être > rayon lysosome)

# ── TrackPy ───────────────────────────────────────────────────────────────
diameter_trackpy   = 11    # diamètre particules (impair) pour tp.locate
search_range       = 4     # portée recherche entre frames (px)
                           # ⚠️  Mettre 2 avec pipeline CNN+Cellpose
memory             = 4     # frames de mémorisation avant abandon trajectoire
threshold_filtered = 20    # longueur minimale trajectoire conservée (frames)


# ═══════════════════════════════════════════════════════════════════════════
# SECTION 3 — CHARGEMENT DES MODÈLES (une seule fois, sur GPU si disponible)
# ═══════════════════════════════════════════════════════════════════════════

# ── CNN SomaCNN ──────────────────────────────────────────────────────────
# map_location=device : charge les poids sur le bon dispositif
# même si le modèle a été entraîné sur un matériel différent
model_cnn = SomaCNN()
model_cnn.load_state_dict(torch.load("soma_cnn_test.pth", map_location=device))
model_cnn = model_cnn.to(device)
model_cnn.eval()   # mode inférence : désactive dropout et batchnorm en mode train

# ── Cellpose cyto3 ───────────────────────────────────────────────────────
# cyto3 : modèle généraliste recommandé pour structures cytoplasmiques
from cellpose import models as cp_models
cellpose_model = cp_models.Cellpose(
    gpu=torch.cuda.is_available(),
    model_type="cyto3"
)
```

## 6. Chargement du stack TIFF

Les images sont stockées sous forme de fichiers TIFF individuels (une frame
par fichier), nommés avec un suffixe `_t<N>` indiquant l'indice temporel.

**Points importants :**
- Le tri est **numérique** (extrait l'entier `N` depuis `_t<N>`) et non
  alphabétique, pour éviter l'ordre incorrect `_t1, _t10, _t100, _t2…`
- Le stack chargé en mémoire a la forme `(T, H, W)` où T = nombre de frames
- Les indices temporels sont extraits et conservés pour la traçabilité
  des figures générées

> **⚠️ Mémoire :** un stack de 200 frames 1024×1024 float32 occupe ~800 Mo.
> Vérifier la RAM disponible avant de charger de larges acquisitions.


```python
# ── Horodatage global (inclut le chargement des modèles) ─────────────────
true_start_time = time.time()

# ── Dossier de sortie ────────────────────────────────────────────────────
# 🔧 Modifier ce chemin pour pointer vers le dossier de sortie souhaité
output_dir = r"C:\Users\huber\Desktop\Analyse_Lysosomes"
os.makedirs(output_dir, exist_ok=True)   # création automatique si inexistant

# ── Chemin des données brutes ─────────────────────────────────────────────
# 🔧 Adapter selon l'environnement (NAS, disque local, etc.)
path = r"\\ICSNAS1\RawData\Arnaud Huber\Lysosomes"

# ── Recherche des fichiers TIFF ───────────────────────────────────────────
# Le motif _t(\d+) capture l'indice temporel dans le nom de fichier
# 🔧 Adapter le motif glob selon la convention de nommage de l'acquisition
files = glob.glob(path + r"\CTRL 1 _ mitotracker_lysotracker_001_t*.tif")

# ── Tri numérique selon l'indice temporel ─────────────────────────────────
# Évite l'ordre alphabétique incorrect : _t1, _t10, _t100, _t2…
files = sorted(files, key=lambda x: int(re.search(r"_t(\d+)", x).group(1)))
print(f"Nombre de fichiers trouvés : {len(files)}")

# ── Chargement en mémoire → stack (T, H, W) ──────────────────────────────
stack = np.array([tiff.imread(f) for f in files])

# ── Vérification du stack ─────────────────────────────────────────────────
print("\n--- Vérification du stack ---")
print(f"Dimensions : {stack.shape}   (attendu : T × H × W)")
print(f"dtype      : {stack.dtype}")
print(f"Mémoire    : {stack.nbytes / 1e6:.1f} Mo")

# Extraction des indices temporels pour la traçabilité des figures
times = [int(re.search(r"_t(\d+)", f).group(1)) for f in files]
```

## 7. Lancement du pipeline de tracking

Appelle `Tracking_Lyso_TrackPy` qui orchestre l'ensemble de l'analyse et
génère automatiquement **7 figures** sauvegardées dans `output_dir` :

| Figure | Description |
|---|---|
| `histogramme_trajectoires[_CNN].png` | Distribution des longueurs de trajectoires |
| `msd_fit[_CNN].png` | MSD moyen + fit loi de puissance (exposant α de diffusion) |
| `nombre_lysosomes_par_frame[_CNN].png` | Évolution du nombre de lysosomes détectés |
| `trajectoires_moyenne[_CNN].png` | Trajectoires superposées sur l'image moyenne |
| `densite_lysosomes[_CNN].png` | Carte de densité spatiale (KDE) |
| `distribution_vitesses[_CNN].png` | Distribution des vitesses instantanées |
| `trajectoires_vitesse[_CNN].png` | Trajectoires colorées par vitesse |

Le suffixe `_CNN` est ajouté si `flag_CNN=True`.

**Interprétation du coefficient de diffusion α (MSD) :**
- α ≈ 1 : diffusion brownienne normale
- α < 1 : diffusion sous-diffusive (lysosome confiné)
- α > 1 : diffusion supra-diffusive (transport actif sur microtubules)


```python
# ── Mesure du temps d'analyse (hors chargement des modèles) ─────────────
start_time = time.time()

# ── Lancement du pipeline complet ────────────────────────────────────────
# flag_CNN=True  → CNN (détection somas) + Cellpose (lysosomes) + TrackPy
# flag_CNN=False → TrackPy seul sur toute l'image (plus rapide, moins précis)
Tracking_Lyso_TrackPy(
    output_dir,
    path,
    stack,
    times,
    pretraitement_sigma,
    clip_limit,
    patch_size,
    cnn_threshold,
    model_cnn,
    cnn_batch_size,
    overlap_thresh,
    cellpose_model,
    cellprob_threshold,
    flow_threshold,
    min_area,
    max_area,
    min_circularity,
    max_axis_ratio,
    top_hat_radius,
    diameter_cellpose,
    diameter_trackpy,
    search_range,
    memory,
    threshold_filtered,
    flag_CNN=True   # 🔧 False = TrackPy seul (plus rapide)
)

# ── Rapport des temps d'exécution ─────────────────────────────────────────
end_time             = time.time()
elapsed_analyse      = end_time - start_time
elapsed_total        = end_time - true_start_time

print(f"\n⏱  Temps d'analyse seule     : {elapsed_analyse:.2f} s")
print(f"⏱  Temps total (+ modèles)   : {elapsed_total:.2f} s")
```
