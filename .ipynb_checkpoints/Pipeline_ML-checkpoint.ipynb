# Pipeline ML — Détection et classification de puncta (somas)

Ce notebook implémente un pipeline complet d'apprentissage automatique pour
la détection et la classification de structures ponctuelles (puncta / somas)
dans des images de microscopie de fluorescence.

**Étapes couvertes :**
1. Construction du dataset de patches à partir de fichiers TIFF
2. Chargement et sous-échantillonnage du dataset
3. Entraînement d'un auto-encodeur convolutionnel (`ConvAutoEncoder`)
4. Réduction dimensionnelle (UMAP) et clustering (KMeans)
5. Sélection manuelle interactive de patches positifs (`PatchSelector`)
6. Entraînement du classificateur CNN (`SomaCNN`)
7. Évaluation (courbe Precision-Recall, seuil optimal)
8. Sauvegarde du modèle

**Dépendances :** toutes les classes et fonctions utilitaires sont définies
dans `Include.ipynb` (chargé ci-dessous).


````python
# Chargement de toutes les dépendances, classes et fonctions utilitaires
# (ConvAutoEncoder, PatchDataset, train_one_epoch, SomaCNN, SomaBinaryDataset,
#  build_dataset_specific_from_folder, detect_candidates, build_blob_dataset, …)
%run "Include.ipynb"
````

## 3. Entraînement de l'auto-encodeur convolutionnel (AE)

L'auto-encodeur (`ConvAutoEncoder`, défini dans `Include.ipynb`) apprend une
représentation latente compacte de chaque patch sans supervision.

**Architecture :**
- Encodeur : 3 couches Conv2d avec stride=2 → espace latent 8×8×64
- Décodeur : 3 couches ConvTranspose2d → reconstruction du patch original

**Critère :** MSELoss entre l'entrée et la reconstruction.

L'espace latent sera ensuite utilisé pour le clustering (UMAP + KMeans).


````python
# ── Hyperparamètres d'entraînement ──────────────────────────────────────────
device           = "cpu"           # 🔧 remplacer par "cuda" si GPU disponible
model            = ConvAutoEncoder().to(device)  # auto-encodeur convolutionnel
optimizer        = optim.Adam(model.parameters(), lr=1e-3)  # optimiseur Adam
criterion        = nn.MSELoss()    # perte de reconstruction pixel à pixel

rounds           = 1               # 🔧 nombre de passes complètes sur le dataset
epochs_per_round = 5               # 🔧 epochs par round
batch_size       = 128             # 🔧 taille de mini-batch

loss_history = []  # historique de la perte pour analyse post-entraînement

# ── Préparation du DataLoader ─────────────────────────────────────────────────
# PatchDataset ajoute la dimension canal (1, H, W) attendue par PyTorch
dataset = PatchDataset(patches)
loader  = DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=0)

print(f"Nombre de patchs total : {len(dataset)}")

# ── Boucle d'entraînement ─────────────────────────────────────────────────────
for r in range(rounds):
    print(f"\n🔁 Round {r+1}/{rounds}")
    for epoch in range(epochs_per_round):
        loss = train_one_epoch(model, loader, optimizer, criterion, device)
        loss_history.append(loss)
        print(f"  Epoch {epoch+1}/{epochs_per_round} | Loss = {loss:.5f}")
````

## 5. Extraction des embeddings et clustering (UMAP + KMeans)

### 5a. Extraction des embeddings
Deux stratégies sont comparées :
- **Flatten** : aplatissement complet de la feature map de l'encodeur → vecteur
  de grande dimension (spatial + global), plus discriminant mais sensible au bruit.
- **GAP** (Global Average Pooling) : moyenne spatiale des feature maps → vecteur
  compact (= nombre de filtres), plus robuste mais perd l'information de position.

### 5b. Réduction UMAP
UMAP projette les embeddings en 2D pour la visualisation. Paramètres clés :
- `n_neighbors=15` : taille du voisinage local (↑ = structure globale plus préservée)
- `min_dist=0.05`  : distance minimale entre points dans l'espace 2D (↓ = clusters plus serrés)

### 5c. Clustering KMeans
KMeans assigne chaque patch à l'un des `n_clusters` groupes. Ajuster `n_clusters`
en fonction du nombre de types de structures attendus dans les données.


````python
# ── Préparation du tenseur d'entrée ─────────────────────────────────────────
# Ajout de la dimension canal : (N, H, W) → (N, 1, H, W)
patches_tensor = torch.tensor(patches, dtype=torch.float32).unsqueeze(1)
dataset        = TensorDataset(patches_tensor)
loader         = DataLoader(dataset, batch_size=128, shuffle=False)

# ── Fonctions d'extraction d'embeddings ──────────────────────────────────────

def extract_embeddings_flatten(model, loader, device="cpu"):
    """Extrait les embeddings en aplatissant la feature map de l'encodeur.
    Retourne un array (N, C*H*W) : riche en information spatiale.
    """
    model.eval()
    Z = []
    with torch.no_grad():
        for batch in loader:
            x      = batch[0].to(device)
            z      = model.encoder(x)           # (batch, C, H, W)
            z_flat = z.view(len(x), -1)         # flatten → (batch, C*H*W)
            Z.append(z_flat)
    return torch.cat(Z, dim=0).cpu().numpy()

def extract_embeddings_gap(model, loader, device="cpu"):
    """Extrait les embeddings par Global Average Pooling.
    Retourne un array (N, C) : compact, invariant à la position spatiale.
    """
    model.eval()
    Z = []
    with torch.no_grad():
        for batch in loader:
            x     = batch[0].to(device)
            z     = model.encoder(x)
            z_gap = z.mean(dim=(2, 3))           # GAP → (batch, C)
            Z.append(z_gap)
    return torch.cat(Z, dim=0).cpu().numpy()

# ── Extraction ───────────────────────────────────────────────────────────────
Z_flat = extract_embeddings_flatten(model, loader)
Z_gap  = extract_embeddings_gap(model, loader)

# ── Réduction UMAP → 2D (pour visualisation) ─────────────────────────────────
reducer_flat  = umap.UMAP(n_neighbors=15, min_dist=0.05, n_components=2, random_state=42)
emb_flat_2d   = reducer_flat.fit_transform(Z_flat)

reducer_gap   = umap.UMAP(n_neighbors=15, min_dist=0.05, n_components=2, random_state=42)
emb_gap_2d    = reducer_gap.fit_transform(Z_gap)

# ── Clustering KMeans ────────────────────────────────────────────────────────
n_clusters  = 20  # 🔧 ajuster selon le nombre de types de structures attendus

kmeans_gap  = KMeans(n_clusters=n_clusters, random_state=42)
labels_gap  = kmeans_gap.fit_predict(Z_gap)

kmeans_flat = KMeans(n_clusters=n_clusters, random_state=42)
labels_flat = kmeans_flat.fit_predict(Z_flat)

# ── Visualisation côte à côte UMAP + clusters ─────────────────────────────────
plt.figure(figsize=(16, 7))

plt.subplot(1, 2, 1)
plt.scatter(emb_gap_2d[:, 0], emb_gap_2d[:, 1], c=labels_gap,
            cmap="tab20", s=5, alpha=0.7)
plt.colorbar(label="Cluster ID")
plt.title("UMAP GAP (global) - Clusters")
plt.xlabel("UMAP1"); plt.ylabel("UMAP2")

plt.subplot(1, 2, 2)
plt.scatter(emb_flat_2d[:, 0], emb_flat_2d[:, 1], c=labels_flat,
            cmap="tab20", s=5, alpha=0.7)
plt.colorbar(label="Cluster ID")
plt.title("UMAP Flatten (spatial + global) - Clusters")
plt.xlabel("UMAP1"); plt.ylabel("UMAP2")

plt.tight_layout()
plt.show()
````

## 6. Inspection visuelle des clusters

`show_patch_clusters` affiche une grille de patches pour chaque cluster, ce qui
permet d'identifier visuellement quels clusters contiennent des somas, des artefacts,
du bruit, etc.

Cette étape est essentielle pour choisir les `soma_clusters` utilisés dans l'étape
d'entraînement du classificateur.


````python
def show_patch_clusters(patches, labels, num_to_show=50):
    """
    Affiche des patchs par cluster pour inspection visuelle.

    Paramètres
    ----------
    patches     : np.array (N, H, W)  — dataset de patches
    labels      : array (N,)           — labels de cluster issus de KMeans
    num_to_show : int                  — nombre de patchs à afficher par cluster
    """
    unique_clusters = np.unique(labels)

    for cluster_id in unique_clusters:
        # Indices des patches appartenant à ce cluster
        cluster_indices = np.where(labels == cluster_id)[0]
        np.random.shuffle(cluster_indices)              # ordre aléatoire pour diversité
        cluster_indices = cluster_indices[:num_to_show] # limiter l'affichage

        n_cols = 10
        n_rows = int(np.ceil(len(cluster_indices) / n_cols))

        plt.figure(figsize=(2 * n_cols, 2 * n_rows))
        plt.suptitle(f"Cluster {cluster_id} | {len(cluster_indices)} patchs", fontsize=16)

        for i, idx in enumerate(cluster_indices):
            plt.subplot(n_rows, n_cols, i + 1)
            plt.imshow(patches[idx], cmap="gray")
            plt.axis("off")

        plt.tight_layout()
        plt.show()

# ── Affichage de tous les clusters (méthode GAP) ──────────────────────────────
# Inspecter les clusters pour identifier lesquels correspondent à des somas
show_patch_clusters(patches, labels_gap, num_to_show=50)
````

## 7. Patchs les plus représentatifs par cluster

Plutôt que des patchs aléatoires, cette fonction sélectionne les patchs les plus
proches du **centroïde** de chaque cluster dans l'espace d'embedding.

Ces patchs sont les exemples "typiques" de chaque cluster, utiles pour valider
la cohérence et l'interprétabilité du clustering.


````python
def plot_representative_patches(patches, embeddings, labels, n_patches=9, cluster_names=None):
    """
    Affiche les patchs les plus proches du centroïde de chaque cluster.

    Paramètres
    ----------
    patches       : array (N, H, W)         — dataset de patches
    embeddings    : array (N, latent_dim)    — embeddings de l'encodeur
    labels        : array (N,)               — labels KMeans
    n_patches     : int                      — nombre de patchs représentatifs par cluster
    cluster_names : list[str] | None         — noms optionnels pour chaque cluster
    """
    clusters = np.unique(labels)

    for c in clusters:
        idx_c  = np.where(labels == c)[0]
        emb_c  = embeddings[idx_c]

        # Calcul du centroïde comme moyenne des embeddings du cluster
        centroid = emb_c.mean(axis=0, keepdims=True)

        # Distance euclidienne de chaque patch au centroïde
        dists       = cdist(emb_c, centroid, metric="euclidean").squeeze()
        closest_idx = idx_c[np.argsort(dists)[:n_patches]]  # patchs les plus proches

        # ── Affichage en grille ───────────────────────────────────────
        n_cols = int(np.ceil(np.sqrt(n_patches)))
        n_rows = int(np.ceil(n_patches / n_cols))
        plt.figure(figsize=(2 * n_cols, 2 * n_rows))

        for i, idx in enumerate(closest_idx):
            plt.subplot(n_rows, n_cols, i + 1)
            plt.imshow(patches[idx], cmap="gray")
            plt.axis("off")

        title = f"Cluster {c}" if cluster_names is None else cluster_names[c]
        plt.suptitle(title)
        plt.tight_layout()
        plt.show()

# ── Exemple d'utilisation avec les embeddings GAP ─────────────────────────────
plot_representative_patches(
    patches    = patches,    # patches 64×64
    embeddings = Z_gap,      # embeddings GAP (N, C)
    labels     = labels_gap, # labels KMeans
    n_patches  = 20
)
````

## 8. Définition du classificateur CNN et fonction d'entrainement

### `SomaBinaryDataset`
Dataset PyTorch pour la classification binaire (soma=1, non-soma=0).
Inclut des augmentations aléatoires (flip horizontal/vertical, rotation ±20°)
appliquées uniquement en mode entraînement pour améliorer la généralisation.

### `train_soma_classifier`

Cette fonction orchestre l'ensemble du cycle d'entraînement du CNN binaire :

1. **Positifs (train)** : patches issus des clusters somas identifiés manuellement
   (`soma_clusters`) ou fournis directement (`manual_pos_idx`).
2. **Négatifs (train)** : sélection aléatoire dans les patches non-positifs,
   avec ratio configurable (`neg_ratio`). Un ratio élevé (~10) est recommandé
   car les somas sont minoritaires dans les images réelles.
3. **Validation** : si `val_pos_idx` est fourni, la perte de validation est
   calculée à chaque époque pour détecter le sur-apprentissage.

> **Note :** `BCELoss` est utilisée car le CNN intègre déjà un `Sigmoid` final.
> Ne pas utiliser `BCEWithLogitsLoss` qui applique sigmoid en interne.


````python
# ===========================
# Dataset PyTorch avec augmentation
# ===========================
class SomaBinaryDataset(Dataset):
    def __init__(self, X, y, augment=False):
        self.X = torch.tensor(X, dtype=torch.float32).unsqueeze(1)
        self.y = torch.tensor(y, dtype=torch.float32).unsqueeze(1)
        self.augment = augment
        # Transformations simples
        self.transforms = T.Compose([
            T.RandomHorizontalFlip(),
            T.RandomVerticalFlip(),
            T.RandomRotation(20),  # ±20°
        ])

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        x, y = self.X[idx], self.y[idx]

        if self.augment:
            # x : (1, H, W) -> transformer nécessite (C,H,W) -> ok
            x = self.transforms(x)

        return x, y

def train_soma_classifier(
    patches,
    labels_gap,
    soma_clusters,
    manual_pos_idx=None,   # indices positifs fournis manuellement (prioritaire sur soma_clusters)
    val_pos_idx=None,      # indices de validation (optionnel)
    batch_size=64,
    epochs=10,
    device="cpu",
    neg_ratio=1,           # 🔧 nb de négatifs par positif (↑ = meilleure précision, ↓ = meilleur rappel)
):
    """Entraîne SomaCNN en classification binaire supervisée.

    Retourne
    --------
    model : SomaCNN entraîné, prêt pour l'inférence
    """
    model     = SomaCNN().to(device)
    criterion = nn.BCELoss()                          # ⚠️ le modèle a déjà un Sigmoid
    optimizer = optim.Adam(model.parameters(), lr=1e-3)

    # ── Positifs d'entraînement ───────────────────────────────────────────────
    if manual_pos_idx is not None:
        # Priorité aux indices fournis manuellement (via PatchSelector)
        train_pos_idx = manual_pos_idx
    else:
        # Fallback : tous les patches appartenant aux clusters somas identifiés
        train_pos_idx = np.where(np.isin(labels_gap, soma_clusters))[0]

    X_pos_train = patches[train_pos_idx]
    y_pos_train = np.ones(len(X_pos_train))

    # ── Négatifs d'entraînement (échantillonnage aléatoire) ──────────────────
    all_idx  = np.arange(len(patches))
    neg_pool = np.setdiff1d(all_idx, train_pos_idx)   # exclure les positifs
    n_neg    = int(len(train_pos_idx) * neg_ratio)
    train_neg_idx = np.random.choice(neg_pool, size=n_neg, replace=False)

    X_neg_train = patches[train_neg_idx]
    y_neg_train = np.zeros(len(train_neg_idx))

    # ── Assemblage du dataset d'entraînement ──────────────────────────────────
    X_train = np.concatenate([X_pos_train, X_neg_train], axis=0)
    y_train = np.concatenate([y_pos_train, y_neg_train], axis=0)

    train_dataset = SomaBinaryDataset(X_train, y_train, augment=True)  # augmentation activée
    train_loader  = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # ── Dataset de validation (optionnel) ─────────────────────────────────────
    if val_pos_idx is not None:
        X_pos_val   = patches[val_pos_idx]
        y_pos_val   = np.ones(len(X_pos_val))
        # Négatifs de validation : exclus des positifs train ET val
        val_neg_idx = np.random.choice(
            np.setdiff1d(all_idx, np.concatenate([train_pos_idx, val_pos_idx])),
            size=len(val_pos_idx), replace=False
        )
        X_val       = np.concatenate([X_pos_val, patches[val_neg_idx]], axis=0)
        y_val       = np.concatenate([y_pos_val, np.zeros(len(val_neg_idx))], axis=0)
        val_dataset = SomaBinaryDataset(X_val, y_val, augment=False)  # pas d'augmentation en val
        val_loader  = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    else:
        val_loader = None

    # ── Boucle d'entraînement ─────────────────────────────────────────────────
    for epoch in range(epochs):
        # ---- Phase train ----
        model.train()
        train_loss = 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            output = model(x)
            loss   = criterion(output, y)
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * len(x)
        train_loss /= len(train_loader.dataset)

        # ---- Phase validation ----
        if val_loader is not None:
            model.eval()
            val_loss = 0
            with torch.no_grad():
                for x, y in val_loader:
                    x, y   = x.to(device), y.to(device)
                    output = model(x)
                    loss   = criterion(output, y)
                    val_loss += loss.item() * len(x)
            val_loss /= len(val_loader.dataset)
            print(f"Epoch {epoch+1}/{epochs} | Train Loss = {train_loss:.5f} | Val Loss = {val_loss:.5f}")
        else:
            print(f"Epoch {epoch+1}/{epochs} | Train Loss = {train_loss:.5f}")

    return model
````

## 9. Sélection manuelle de patches positifs (`PatchSelector`)

Interface interactive matplotlib (fenêtre externe, `%matplotlib tk`) permettant
de valider manuellement des patches comme exemples positifs (somas) :

- **Clic gauche** sur un patch : bascule la sélection (coché ✓ en vert = sélectionné)
- **← / →** : naviguer entre les blocs de patches
- **S** : sauvegarder la sélection courante dans `manual_soma_selection.npy`
- **Q** : fermer la fenêtre et afficher le résumé

> Cette cellule nécessite un backend graphique (`%matplotlib tk`).
> En environnement sans affichage (serveur SSH, Jupyter Lab distant), utiliser
> `%matplotlib inline` et adapter la sélection.


````python
%matplotlib tk
import matplotlib.pyplot as plt
import numpy as np

class PatchSelector:
    """Interface graphique interactive pour la sélection manuelle de patches.

    Affiche les patches par blocs de `block_size` dans une grille `grid_shape`.
    Permet de sélectionner/désélectionner des patches au clic, de naviguer
    avec les flèches clavier, et de sauvegarder la sélection.
    """

    def __init__(self, patches, block_size=50, grid_shape=(5, 10)):
        self.patches          = patches
        self.block_size       = block_size
        self.grid_shape       = grid_shape
        self.current_block    = 0
        self.selected_global  = set()  # indices globaux des patches sélectionnés

        # Initialisation de la figure matplotlib
        self.fig, self.axes = plt.subplots(
            grid_shape[0], grid_shape[1], figsize=(12, 6)
        )
        self.axes = self.axes.flatten()

        # Connexion des événements souris et clavier
        self.fig.canvas.mpl_connect("button_press_event", self.on_click)
        self.fig.canvas.mpl_connect("key_press_event",    self.on_key)

        self.update_display()
        plt.tight_layout()
        plt.show()

    def update_display(self):
        """Rafraîchit la grille avec le bloc courant de patches."""
        start = self.current_block * self.block_size
        end   = min(start + self.block_size, len(self.patches))
        self.current_indices = list(range(start, end))

        for i, ax in enumerate(self.axes):
            ax.clear()
            if i < len(self.current_indices):
                idx   = self.current_indices[i]
                title = str(idx)
                ax.imshow(self.patches[idx], cmap="gray")
                if idx in self.selected_global:
                    title += " ✓"
                    ax.set_title(title, color="green", fontsize=8)
                else:
                    ax.set_title(title, fontsize=8)
            ax.axis("off")

        self.fig.canvas.draw_idle()

    def on_click(self, event):
        """Bascule la sélection du patch cliqué."""
        for i, ax in enumerate(self.axes):
            if ax == event.inaxes and i < len(self.current_indices):
                idx = self.current_indices[i]
                if idx in self.selected_global:
                    self.selected_global.remove(idx)  # désélectionner
                else:
                    self.selected_global.add(idx)     # sélectionner
                self.update_display()
                break

    def on_key(self, event):
        """Gestion des raccourcis clavier."""
        if event.key == "right":
            # Bloc suivant (si disponible)
            if (self.current_block + 1) * self.block_size < len(self.patches):
                self.current_block += 1
                self.update_display()
        elif event.key == "left":
            # Bloc précédent
            if self.current_block > 0:
                self.current_block -= 1
                self.update_display()
        elif event.key == "q":
            # Quitter et afficher le résumé
            print("\n===== FIN SÉLECTION =====")
            print("Nombre total sélectionné :", len(self.selected_global))
            print("Indices :", sorted(self.selected_global))
            plt.close(self.fig)
        elif event.key == "s":
            # Sauvegarde intermédiaire
            np.save("manual_soma_selection.npy",
                    np.array(sorted(self.selected_global)))
            print("Sélection sauvegardée.")
````

## 13. Inférence : prédiction des probabilités sur tout le dataset

`predict_probs` exécute le CNN en mode évaluation (sans gradient) sur l'ensemble
des patches et retourne un tableau de probabilités ∈ [0, 1].

Ces probabilités seront utilisées pour :
- Tracer la courbe Precision-Recall
- Choisir le seuil de décision optimal
- Visualiser la distribution des scores


````python
def predict_probs(model, patches, batch_size=128, device="cpu"):
    """Calcule les probabilités CNN pour tous les patches.

    Paramètres
    ----------
    model      : SomaCNN entraîné
    patches    : np.array (N, H, W)
    batch_size : int — taille de batch pour l'inférence
    device     : str — dispositif de calcul

    Retourne
    --------
    probs : np.array (N,) float — probabilité soma pour chaque patch
    """
    model.eval()
    # Utilisation de SomaBinaryDataset avec des labels factices (non utilisés en inférence)
    dataset = SomaBinaryDataset(patches, np.zeros(len(patches)))
    loader  = DataLoader(dataset, batch_size=batch_size, shuffle=False)
    probs   = []
    with torch.no_grad():
        for x, _ in loader:
            x      = x.to(device)
            y_pred = model(x)
            probs.append(y_pred.cpu())
    return torch.cat(probs, dim=0).numpy().flatten()
````

## 14. Évaluation : courbe Precision-Recall et seuil optimal

### Construction du jeu de validation
Un jeu de validation équilibré est construit avec :
- tous les positifs de validation (`val_pos_idx`)
- 10× plus de négatifs (pour simuler le déséquilibre réel)

### Choix du seuil
Deux stratégies de sélection du seuil de décision sont comparées :

1. **Seuil F1 optimal** : maximise F1 = 2·P·R / (P+R) → compromis précision/rappel
2. **Seuil precision contrainte** : recall maximal sous contrainte P ≥ 90 %
   → préférable si les faux positifs sont coûteux (annotations manuelles, etc.)


````python
from sklearn.metrics import precision_recall_curve, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import numpy as np

# ── Inférence sur tout le dataset ─────────────────────────────────────────────
print(f"Nombre total de patches : {len(patches)}")
probs = predict_probs(model, patches)

# ── Construction du jeu de validation ────────────────────────────────────────
# y_true : 1 pour les positifs validés, 0 pour tous les autres
y_true   = np.zeros(len(patches))
y_true[val_pos_idx] = 1
all_idx  = np.arange(len(patches))

# Pool de négatifs : exclure les positifs d'entraînement
neg_pool    = np.setdiff1d(all_idx, train_pos_idx)
val_neg_idx = np.random.choice(neg_pool, size=10 * len(val_pos_idx), replace=False)
print(f"Négatifs de validation : {len(val_neg_idx)}")

val_indices  = np.concatenate([val_pos_idx, val_neg_idx])
y_true_val   = y_true[val_indices]
probs_val    = probs[val_indices]
print(f"Taille du jeu de validation : {len(y_true_val)}")

# ── Courbe Precision-Recall ───────────────────────────────────────────────────
precision_curve, recall_curve, thresholds = precision_recall_curve(y_true_val, probs_val)

plt.plot(recall_curve, precision_curve)
plt.xlabel("Recall"); plt.ylabel("Precision")
plt.title("Courbe Precision-Recall (jeu de validation)")
plt.show()

# ── Seuil 1 : F1 optimal ─────────────────────────────────────────────────────
f1_scores    = 2 * (precision_curve * recall_curve) / (precision_curve + recall_curve + 1e-8)
best_idx_f1  = np.argmax(f1_scores)
best_threshold_f1 = thresholds[best_idx_f1]
print(f"Seuil optimal (F1) : {best_threshold_f1:.4f} | F1 = {f1_scores[best_idx_f1]:.4f}")

# ── Seuil 2 : Precision ≥ 90 % avec recall maximal ───────────────────────────
target_precision = 0.9
valid_idx        = np.where(precision_curve[:-1] >= target_precision)[0]

if len(valid_idx) > 0:
    best_idx_prec      = valid_idx[np.argmax(recall_curve[valid_idx])]
    best_threshold_prec = thresholds[best_idx_prec]
    print(f"Seuil (Precision ≥ {target_precision*100:.0f}%) : {best_threshold_prec:.4f}")
else:
    best_threshold_prec = None
    print("⚠️  Precision cible non atteinte")

# ── Évaluation comparative des deux seuils ────────────────────────────────────
def evaluate_threshold(threshold, name):
    """Affiche precision, recall et F1 pour un seuil donné."""
    if threshold is None:
        return
    y_pred = (probs_val >= threshold).astype(int)
    print(f"\n--- {name} ---")
    print(f"Threshold  : {threshold:.4f}")
    print(f"Precision  : {precision_score(y_true_val, y_pred):.4f}")
    print(f"Recall     : {recall_score(y_true_val, y_pred):.4f}")
    print(f"F1         : {f1_score(y_true_val, y_pred):.4f}")

evaluate_threshold(best_threshold_f1,   "F1 optimal")
evaluate_threshold(best_threshold_prec, "Precision contrainte")
````

## 15. Distribution des probabilités CNN

L'histogramme (échelle logarithmique en Y) permet de visualiser la séparation
entre somas (probabilités élevées) et non-somas (probabilités basses).

Une bonne séparation bimodale indique que le CNN discrimine bien les deux classes.
Les seuils choisis à l'étape précédente sont affichés en surimpression.


````python
import matplotlib.pyplot as plt

print(f"Nombre de patches scorés : {len(probs)}")

plt.figure(figsize=(7, 5))
plt.hist(probs, bins=20, color="skyblue", edgecolor="k")

# Affichage des deux seuils candidats
plt.axvline(best_threshold_f1,
            color="red", linestyle="--",
            label=f"Seuil F1 optimal = {best_threshold_f1:.2f}")
if best_threshold_prec is not None:
    plt.axvline(best_threshold_prec,
                color="cyan", linestyle="--",
                label=f"Seuil precision 90% = {best_threshold_prec:.2f}")

plt.xlabel("Probabilité CNN")
plt.ylabel("Nombre de patches")
plt.title("Distribution des probabilités CNN sur tout le dataset")
plt.yscale("log")   # log pour mieux visualiser les queues
plt.grid(True)
plt.legend()
plt.show()

# ── Statistiques descriptives ─────────────────────────────────────────────────
print("===== Statistiques des probabilités =====")
print(f"Min     : {np.min(probs):.4f}")
print(f"Max     : {np.max(probs):.4f}")
print(f"Moyenne : {np.mean(probs):.4f}")
print(f"Médiane : {np.median(probs):.4f}")
````
